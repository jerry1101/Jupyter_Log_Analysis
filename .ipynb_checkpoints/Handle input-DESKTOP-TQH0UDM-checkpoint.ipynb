# Prepare raw data for analysis

```python
%%bash
#cd  \\\\cs-p-log1\\ProductionLogging\\2019-05\\Info\\Services\\10.24.87.51\\
cd  \\\\cs-p-log1\\ProductionLogging\\2019-05\\Info\\Services
pwd
va="5_31_2019.log"
find . -name "$va" -exec grep -H 'GetSkus' {} \; > c://tmp//a2.txt
```

```python
%%bash -s "c://tmp//a2.txt"
echo 'Nos of getSkus'
grep 'Parameters' "$1" | wc -l
```

```python
%%bash -s "c://tmp//a2.txt"
echo 'Nos of getSkus W/ one item'
grep 'Parameters'  "$1"| grep -v ',' |wc -l
grep -v "Result" "$1"| grep -E -A1 "(ItemKeys=[0-9]{4,6}\]){1}" | grep -v "\--" | grep -v 'Parameters'
```

```python
%%bash -s "c://tmp//a2.txt"
echo 'Nos of getSkus W/ over ten item'
grep 'Parameters'  "$1" | grep -E "([0-9]{4,6},){10}"| wc -l
grep -v "Result" "$1"| grep -E -A1 "([0-9]{4,6},){10}"| grep -v "\--"| grep -v 'Parameters'
```

```python
%%bash -s "c://tmp//a2.txt"
echo 'Nos of getSkus W/ two to ten item'
grep 'Parameters'  "$1" | grep -E -v "([0-9]{4,6},){10}" | grep "," | wc -l
grep -v "Result" "$1"| grep -E -v "([0-9]{4,6},){10}"| grep -A1 "," | grep -v "\--"| grep -v 'Parameters'
```

# Start data reformat

```python
%%bash 
file="RawData_5_01_2019.txt"
grep -v 'Result' "$file" | grep -v ',' | grep 'Parameters' -A1 | grep -v "\--" | grep -v 'Para'| sed 's/GetSkus://g' | sed 's/ : /|/g'| sed 's/  //g' > t1.txt
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df=pd.read_csv("t1.txt",sep='|',names=['datetime','server','ms'])

#df['strdatetime']=df['t1'].astype(str) +':'+df['t2'].astype(str) +':'+ df['t3']
#df['datetime']=pd.to_datetime(df['strdatetime'])
#newdf= df[['datetime','server','ms']]
df.head()
```

# Analysis

```python
perc =[.20, .40, .60, .80,.95] 
df.describe(percentiles= perc)
```

```python
plt.figure(figsize=(20,8),dpi=300)
plt.hist(df['ms'], bins=30)
plt.ylabel('#');
plt.xlabel('Run Time (ms)');
plt.title('GetSkus() Run Time Distribution on 5/1')
plt.grid(True)
```
