## 1. Init __logger__

```python
import glob
import logging
import logging.handlers

LOG_FILENAME = 'local.log'

# Set up a specific logger with our desired output level
local_logger = logging.getLogger('local')
local_logger.setLevel(logging.INFO)

# Add the log message handler to the logger
handler = logging.handlers.RotatingFileHandler(
              LOG_FILENAME, maxBytes=20, backupCount=5)

local_logger.addHandler(handler)
```

## 2. Import history data
### 2.1 use excel file as data source

```python
import numpy as np
import pandas as pd
from pprint import pprint
df=pd.read_excel('GetSupplyChain.xlsx', index_col=0)  
df.head()
from bs4 import BeautifulSoup


def add_parameter(existing_dic, key, value):
    if (value != ''):
        existing_dic.update([(key,value)])
        return True
    return False
    
    

history_parameters= []
for message in df['SoapMessage']:
    soup = BeautifulSoup(message[1:],'xml')
    items = soup.find_all('ItemKey')
    srs = soup.find_all('SupplyRouteKey')
    dls = soup.find_all('DestinationLocationKey')
    wscs = soup.find_all('WithSupplyChainDetails')
    os = soup.find_all('OriginLocationKey')
    #print(' '.join(['itemKey:',len(items[0].get_text()),'SupplyRouteKey:',len(srs[0].get_text())]))
    #print('|'.join([os[0].get_text(),dls[0].get_text(),srs[0].get_text(),items[0].get_text(),wscs[0].get_text()]))
    dic = {}
    add_parameter(dic,'OriginLocationKey',os[0].get_text())
    add_parameter(dic,'DestinationLocationKey',dls[0].get_text())
    add_parameter(dic,'SupplyRouteKey',srs[0].get_text())
    add_parameter(dic,'ItemKey',items[0].get_text())
    add_parameter(dic,'WithSupplyChainDetails',wscs[0].get_text())
    history_parameters.append(dic)
#     history_parameters.append(dict(OriginLocationKey=os[0].get_text(),
#                                   DestinationLocationKey=dls[0].get_text(),
#                                   SupplyRouteKey=srs[0].get_text(),
#                                   ItemKey=items[0].get_text(),
#                                   WithSupplyChainDetails=wscs[0].get_text()))
pprint(history_parameters)
```
